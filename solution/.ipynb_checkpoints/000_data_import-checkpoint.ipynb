```python
import pandas as pd
from   matplotlib import pyplot as plt
import numpy as np
from   statsmodels.tsa import stattools
import matplotlib
from   sklearn.metrics import mean_absolute_error
from   keras import backend as K
from   keras.layers import *
from   keras.models import *
from   keras.callbacks import *
import pickle
```

```python
user = {}
missing_val = []
user_no = [i for i in range(1,23)]
```

```python
for i in range(1,23):
    file_name = './data/'+str(i)+'.csv'
    user[i] = pd.read_csv(file_name,header=None)
    user[i].columns = ['time_step','x_accelet','y_accelet','z_accelet']
    missing_val.append(sum(user[i].isnull().sum()))
```

```python
plt.plot(user_no,missing_val)
plt.xlabel('User No')
plt.ylabel('Missing Numbers')
plt.savefig('missing_numbers.png')
```

# Conclusion - No Missing Value

```python
for i in range(1,23):
    user[i]['total_accelet'] = user[i].apply(lambda row: np.sqrt(row['x_accelet']**2+row.loc['y_accelet']**2+row.loc['z_accelet']**2),axis=1)
```

# Number of Data Points

```python
data_point_count = []
for i in range(1,23):
    data_point_count.append(user[i].shape[0])
```

```python
plt.barh(np.arange(1,23),data_point_count)
plt.xlabel('Number of data points')
plt.ylabel('User number')
```

```python
matplotlib.rcParams['figure.figsize'] = (20, 5)
k = 0
nCol        = 22
nRow        = 8
graph_range = [131,132,133]

for i in range(nRow+1):
    start = 3*i
    for j in range(3):
        plt.subplot(graph_range[j])
        if start+j+1<=22:
            plt.plot(user[start+j+1]['time_step'],user[start+j+1]['total_accelet'])
            plt.xlabel('time_step')
            plt.ylabel('total_accelet')
            title_name = 'User Acceleration ' + str(start+j+1)
            plt.title(title_name)
    plt.show()
```

### Anomaly Detection among Users

#### User 3,5 and 19 has less data points and are behaving erratically.
#### Therefore removing from domain

### Data Smoothening

```python
for i in range(1,23):
    if i not in [3,5,19]:
        user[i]['total_accelet'] = user[i]['total_accelet'].rolling(window=5).mean()
        user[i].dropna(inplace=True)
        user[i].reset_index(drop=True)
```

```python
with open('./data/user.pickle', 'wb') as handle:
    pickle.dump(user, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
