```python
import pandas as pd
from   matplotlib import pyplot as plt
import numpy as np
from   statsmodels.tsa import stattools
import matplotlib
from   sklearn.metrics import mean_absolute_error
from   keras import backend as K
from   keras.layers import *
from   keras.models import *
from   keras.callbacks import *
import pickle
```

```python
with open('./data/user.pickle', 'rb') as handle:
    user = pickle.load(handle)
```

### Understanding Temporal Structures

### Window Size=100 with overlap of 50

#### Function for defining the window on data

```python
def window(axis,win=100):
    start = 0;
    size = axis.count();
    while (start < size):
        end = start + win
        yield start,end
        start = start+win//2
```

### Features which are extracted from Raw sensor data

```python
def window_summary(axis, start, end):
    acf = stattools.acf(axis[start:end])
    avg = axis[start:end].mean()
    std = axis[start:end].std()
    min_iter = axis[start:end].min()
    max_iter = axis[start:end].max()
    max_min = max_iter - min_iter
    return acf.mean(),avg,std,min_iter,max_iter,max_min
```

```python
acf = []
avg = []
std = []
min_com = []
max_com = []
max_min = []
graph_range = [161,162,163,164,165,166]
user_features = {}
```

### Feature Visualization

```python
matplotlib.rcParams['figure.figsize'] = (20, 5)
for i in range(1,23):
    if i not in [3,5,19]:
        
        window_no = [0]
        acf_user  = []
        avg_user  = []
        std_user  = []
        min_user  = []
        max_user  = []
        max_min_user = []

        for (start, end) in window(user[i]['time_step']):
            window_no_index = window_no[-1]+1
            window_no.append(window_no_index)
            acf_mean_temp,avg_temp,std_temp,min_iter_temp,max_iter_temp,max_min_temp = window_summary(user[i]['total_accelet'],start, end)
            acf_user.append(acf_mean_temp)
            avg_user.append(avg_temp)
            std_user.append(std_temp)
            min_user.append(min_iter_temp)
            max_user.append(max_iter_temp)
            max_min_user.append(max_min_temp)


        plt.subplot(graph_range[0])    
        plt.plot(window_no[1:],acf_user)
        plt.xlabel('window number')
        plt.ylabel('Acf')
        title_name = 'User ' + str(i)
        plt.title(title_name)
        
        
        plt.subplot(graph_range[1])    
        plt.plot(window_no[1:],avg_user)
        plt.xlabel('window number')
        plt.ylabel('Average Acceleration')
        title_name = 'User ' + str(i)
        plt.title(title_name)
        
        
        plt.subplot(graph_range[2])    
        plt.plot(window_no[1:],std_user)
        plt.xlabel('window number')
        plt.ylabel('Standard Deviation')
        title_name = 'User ' + str(i)
        plt.title(title_name)
        
        
        plt.subplot(graph_range[3])    
        plt.plot(window_no[1:],min_user)
        plt.xlabel('window number')
        plt.ylabel('Minimum of Acceleration')
        title_name = 'User ' + str(i)
        plt.title(title_name)
        
        
        plt.subplot(graph_range[4])    
        plt.plot(window_no[1:],max_user)
        plt.xlabel('window number')
        plt.ylabel('Maximum of Acceleration')
        title_name = 'User ' + str(i)
        plt.title(title_name)
        
        
        plt.subplot(graph_range[5])    
        plt.plot(window_no[1:],max_min_user)
        plt.xlabel('window number')
        plt.ylabel('Difference of Max and Min')
        title_name = 'User ' + str(i)
        plt.title(title_name)
        
        plt.show()

        acf.extend(acf_user)
        avg.extend(avg_user)
        std.extend(std_user)
        min_com.extend(min_user)
        max_com.extend(max_user)
        max_min.extend(max_min_user)
        
        features = pd.DataFrame({'acf':acf_user,'avg':avg_user,'std':std_user,'min':min_user,'max':max_user,'max_min':max_min_user})
        user_features[i] = features
```

```python
with open('./data/user_features.pickle', 'wb') as handle:
    pickle.dump(user_features, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
